<h1>Color Classification using Histogram Color Classifier</h1>

In this example we train a simple algorithm to classify images using histogram color classifier. For a detailed explanation visit this <a href="https://mpatacchiola.github.io/blog/2016/11/12/the-simplest-classifier-histogram-intersection.html">link</a>.

```python
#Import required libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from deepgaze.color_classification import HistogramColorClassifier
```

Define classifier instance 

```python
my_classifier = HistogramColorClassifier(channels=[0, 1, 2], hist_size=[128, 128, 128], hist_range=[0, 256, 0, 256, 0, 256], hist_type='BGR')
```

Read image instances we would want to classify 

```python
model_1 = cv2.imread('model_1.png') #Flash
model_2 = cv2.imread('model_2.png') #Batman
model_3 = cv2.imread('model_3.png') #Hulk
model_4 = cv2.imread('model_4.png') #Superman
model_5 = cv2.imread('model_5.png') #Capt. America
model_6 = cv2.imread('model_6.png') #Wonder Woman
model_7 = cv2.imread('model_7.png') #Iron Man
model_8 = cv2.imread('model_8.png') #Wolverine
```

Train the classifier instance with the read images

```python
my_classifier.addModelHistogram(model_1,name="Flash")
my_classifier.addModelHistogram(model_2,name="Batman")
my_classifier.addModelHistogram(model_3,name="Hulk")
my_classifier.addModelHistogram(model_4,name="Superman")
my_classifier.addModelHistogram(model_5,name="Captain America")
my_classifier.addModelHistogram(model_6,name="Wonder Woman")
my_classifier.addModelHistogram(model_7,name="Iron Man")
my_classifier.addModelHistogram(model_8,name="Wolverine")
```

```python
image=cv2.imread('image_2.jpg')
```

```python
plt.imshow(image)
plt.show()
```

```python
comparison_array,name_list = my_classifier.returnHistogramComparisonArray(image, method="intersection")
comparison_distribution = comparison_array / np.sum(comparison_array)
```

```python
print("Comparison Array:")
print(dict(zip(name_list, comparison_array)))
```

```python
print("Distribution Array: ")
print(dict(zip(name_list,comparison_distribution)))
```

```python
plt.hist(dict(zip(name_list,comparison_distribution)))
plt.show()
```
